## Field Dependence of the Resonant Absorption Frequency

Using a distilled water sample, the audio oscillator was set to an 80 Hz triangle wave at maximum amplitude. The RF Frequency was set near 14.29 MHz. 

The block diagram is shown below:

![](nmr_1.png)


At room temperature, the number of protons in the lower energy level ($N^+$) is slightly greater than the higher energy level ($N^-$). The proportion is given by a Boltzmann distribution:

$$\frac{N^-}{N^+} = e^{\frac{-E}{kT}}$$

where $E= h \gamma B$ is the energy between states.

The oscilloscope displays two peaks at resonance frequency since we are performing a linear sweep forwards and backwards.

```python
dataframe = pd.read_csv("resonance_vs_field.csv")
dataframe
```

```python
res_freq = dataframe['Res_Freq'] / 1e6
b_field = dataframe['B_Field']
plt.xlabel("Resonance Frequency (MHz)")
plt.ylabel("Magnetic Field (kG)")
plt.plot(res_freq, b_field, 'bo')
plt.grid(True)

m, b = np.polyfit(res_freq, b_field, 1)
plt.plot(res_freq, m*res_freq + b, '-')
print("Gyromagnetic Ratio:", 1/m * (2*np.pi), "MHz/kG")
```

Note that there is a linear relationship between the magnetic field strength and resonance frequency (allowing for a tolerance of $\pm 0.1 kG$ and $\pm 200 Hz$ respectively). The inverse slope of this line yields the gyromagnetic ratio $\gamma$ for $^1{H}$ since,

$$E = h \gamma B \text{ and } E = h \nu \implies$$

$$ \nu = \gamma B \text{ where } \omega = 2 \pi \nu$$

and so $\gamma = 29.02 \cdot 10^7\, \text{Rad} \, s^{-1} T^{-1}$, which is comparable to the tabulated value $26.75 \cdot 10^7 \, \text{Rad} \, s^{-1} T^{-1}$.

The current in the Helmholtz coils (by using the 1 Ohm resistor in series with the coils) can be calculated in the following manner:

$$V = IR$$
$$188 \text{ mV} = I(1 \Omega)$$

The formula for the magnetic field produced by the coils is:

$$ B = \bigg(\frac{4}{5}\bigg)^\frac{3}{2} \frac{\mu_0 n I}{R} $$

where $n$ is the number of turns of the coil, $R$ is the radius of the coils (equal to the separation distance between solenoids), and $\mu_0$ is the magnetic permeability. So,

$$ B = \bigg(\frac{4}{5}\bigg)^\frac{3}{2} \frac{ (4 \pi \cdot 10^{-7} H \cdot m^{-1})(200)(0.188 A) }{0.0175 m} = 0.001932 \,T = 19.32 \,G$$

This value represents roughly 8.8 divisions on the oscilloscope. Since the sample's linewidth (FWHM) spanned roughly 0.4 divisions, the linewidth is ~0.9 Gauss.

The lock-in amplifier detects only signals in phase with the reference and improves the signal to noise ratio.

### Lock-in Amplifier

<ul>
    <li> Phase-sensitive detection </li>
    <li> Low-pass filter </li>
</ul>
![](nmr_3.png)

$$V_{psd} = V_{sig} \sin{(2\pi f_{s} + \theta_{sig})} V_{ref} \sin{(2\pi f_{r} + \theta_{ref})} = $$
$$\frac{1}{2} V_{sig}V_L \sin{(2\pi[f_{s}-f_{r}] + \theta_{sig} - \theta_{ref})} - \frac{1}{2} V_{sig}V_L \sin{(2\pi[f_{s}+f_{r}] + \theta_{sig} + \theta_{ref})}$$

### Linewidth vs Modulation Frequency

<ul>
    <li>The modulation amplitude was set by the variable attenuator.</li>
    <li>The output of the lock-in amplifier is the derivative of the absorption profile.</li>
    <li>The linewidth can be measured by looking at the derivative curve's peaks.</li>
</ul>

<img src="nmr_4.png" style="width: 380px">

```python
DATA_SETS = 11
linewidths = np.empty([11,])
sweeps = np.empty([11,])
fig = plt.figure(figsize=(15,20))
for i in range(1, DATA_SETS + 1):
    axis = fig.add_subplot(3, 4, i)
    df = pd.read_csv("nmr-data/WL_NMR_" + str(i) + ".csv")
    data = df[2:]
    axis.plot(data) 
    sweep = df.iloc[1][0] - df.iloc[0][0]
    sweeps[i - 1] = sweep
    freq_per_point = sweep / (len(df.index) - 2)
    linewidths[i - 1] = (data.idxmax() - data.idxmin()) * freq_per_point
```

```python
modulations = np.array([8.6, 7.8, 7.0, 6.2, 5.4, 4.6, 3.8, 3.0, 2.2, 1.4, 0.6]) / 8.6
rel_widths = modulations * linewidths / sweeps
print(rel_widths)
plt.plot(modulations, rel_widths, 'bo')
```

Note that the relative linewidth generally decreases with smaller modulation frequencies and then begins to level off at around 0.01. However, there are clearly some errors, as shown by the inconsistencies near high and low modulations.

#### Varying the Samples

The following are derivative curves for glycerin, methanol, and water. The integral (absoption profile) is shown as well.

```python
linewidths = np.empty([3,])
res_freq_avgs = np.empty([3,])
fig = plt.figure(figsize=(15,3))
x = np.empty([3, 200])
y = np.empty([3, 200])
i = 1
for filename in ["glycerin_2", "methanol_1", "water"]:
    axis = fig.add_subplot(1, 3, i)
    df = pd.read_csv("nmr-data/" + filename + ".csv")
    data = df[2:]
    
    start = df.iloc[0][0]
    stop = df.iloc[1][0]
    sweep = stop - start
    freq_per_point = sweep / (len(df.index) - 2)
    x_divs = np.arange(df.iloc[0][0], df.iloc[1][0], freq_per_point)
    linewidths[i-1] = (data.idxmax() - data.idxmin()) * freq_per_point
    axis.plot(x_divs, data)
    
    sample = df[int(data.idxmin()): int(data.idxmax())]
    sample = np.where(sample <= 0)
    sample_freq_pt = int(data.idxmin()) + sample[0][-1] 
    res_freq_avg = start + (2 * sample_freq_pt + 1) * freq_per_point / 2
    res_freq_avgs[i-1] = res_freq_avg
    x[i-1] = x_divs
    y[i-1] = data.T
    i += 1

print("Glycerin:", linewidths[0], "Hz")
print("Methanol:", linewidths[1], "Hz")
print("Water:", linewidths[2], "Hz")
```

```python
fig = plt.figure(figsize=(15,3))
y_ints = np.empty([3, 200])
for i in range(1, 4):
    y_int = integrate.cumtrapz(y[i-1], x[i-1], initial=0)
    y_ints[i-1] = y_int
    axis = fig.add_subplot(1, 3, i)
    axis.plot(x[i-1], y_int)
```

Here the plots are overlayed to compare the chemical shift between substances. The corresponding resonance frequencies are estimated by averaging the two frequencies closest to zero on the derivative plot.

```python
plt.subplot()
plt.plot(x[0],y[0],label="Glycerin")
plt.plot(x[1],y[1],label="Methanol")
plt.plot(x[2],y[2],label="Water")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
i = 0
plt.subplot()
for sample in ["Glycerin", "Methanol", "Water"]:
    print(sample, res_freq_avgs[i], "Hz")
    plt.plot(x[i], y_ints[i], label=sample)
    plt.axvline(x=res_freq_avgs[i])
    i += 1

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

#### Investigating Time Constant $T_2$

The audio frequency was set to 80 Hz and maximum amplitude. A saturated LiF water sample was used. We extrapolate T2 by fitting an exponential curve to the peaks.

![](TEK0002.jpg)

```python
df = pd.read_csv("TEK0001.csv")
df.head()
```

```python
y = df['y'] + 0.01 #zero offset
x = df['x']
indexes = find_peaks_cwt(y, np.arange(1, 120))
indexes = np.array(indexes)
print("Peak detect indices:", indexes)
plt.plot(x, y)
plt.plot(x[indexes], y[indexes],'ro')
```

The exponential will be fit to the first four peaks. To fit a curve of the form $y=Ae^{Bx}$, we take the logarithm of both sides so that $\log{y}=\log{A}+Bx$. Then, simply fit $\log{y}$ against $x$.

```python
adj_index = indexes[5:9]
min_val = y[adj_index[-1]]
selected_x = x[adj_index]
selected_y = y[adj_index]
log_y = np.log(selected_y + 0.01)
plt.plot(selected_x, log_y, 'bo')
m, b = np.polyfit(selected_x, log_y, 1)
plt.plot(selected_x, m*selected_x + b, '-')
print("m:", m)
print("b:", b)
```

Since, $M_{xy} = M_0 e^{-\frac{t}{T_2}}$, this means that $\frac{-1}{T_2} = -1173.61$ and so $T_2 = 0.85 \, ms$
